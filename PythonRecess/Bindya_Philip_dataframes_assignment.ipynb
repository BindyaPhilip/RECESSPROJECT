# DataFrame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1.Reading the file
df = pd.read_csv('AfricaCupofNationsMatches.csv')
df
```

```python
# 2.the first 7 rows of the dataset
df.head(7)
```

```python
#3.Displaying the HomeTeam','AwayTeam','HomeTeamGoals','AwayTeamGoals columns only
df[['HomeTeam','AwayTeam','HomeTeamGoals','AwayTeamGoals']]
```

```python
#4.Selecting the rows where Egypt appears.
selected_rows = df[df.applymap(lambda x: 'Egypt' in str(x)).any(axis=1)]
selected_rows.head(10)
```

```python
#5.Counting the number of rows and columns 
number_rows = len(df)
number_columns = len(df.columns)

print("Number of rows:", number_rows)
print("Number of columns:", number_columns)
```

```python
#6.Selecting the rows where the ‘Attendance’ is missing
missing_attendance = df[df['Attendance'].isna()]

missing_attendance
```

```python
#7.Selecting the rows where the ‘HomeTeamGoals’ are between 3 and 6 inclusive.
selected_row = df[(df['HomeTeamGoals'] >= 3) & (df['HomeTeamGoals'] <= 6)]
selected_row.head(15)
```

```python
#8.Changing the “AwayTeamGoals” in the 3rd row to 10.
df.loc[2, 'AwayTeamGoals'] = 10
df
```

```python
#9.Sorting the home team column in ascending order and the HomeTeamGoals in descending order
sorted_df = df.sort_values(by=['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])
sorted_df
```

```python
#10.Getting list from DataFrame column headers.
df.columns.tolist()
```

```python
#11.creating a new column that will be have the summation of total goals scored by the home team and away team on a particular match
df['TotalGoals'] = df['HomeTeamGoals'] + df['AwayTeamGoals']
df.head(10)
```

```python
#12.Creating two new rows
new_row1 = {'Year': 2022, 'Date': '2022-01-01', 'Time': '15:00', 'HomeTeam': 'Uganda', 'AwayTeam': 'Kenya',
             'HomeTeamGoals': 2, 'AwayTeamGoals': 3, 'Stage': 'Group Stage', 'SpecialWinConditions': '',
             'Stadium': 'Namboole', 'City': 'Kampala', 'Attendance': 50000, 'TotalGoals': 5}
new_row2 = {'Year': 2022, 'Date': '2022-02-01', 'Time': '16:00', 'HomeTeam': 'Rwanda', 'AwayTeam': 'Tanzania',
             'HomeTeamGoals': 2, 'AwayTeamGoals': 3, 'Stage': 'Group Stage', 'SpecialWinConditions': '',
             'Stadium': 'Kigali stadium', 'City': 'Kigali', 'Attendance': 70000, 'TotalGoals': 5}
df.loc[len(df)] = new_row1
df.loc[len(df)] = new_row2
df
```

```python
#13.Changing the country ‘Uganda’ to ‘China’ in the ‘AwayTeam’ column of the DataFrame.
df['AwayTeam'] = df['AwayTeam'].str.strip().replace('Uganda','China')
df.head(20)
```

```python
#14.Resetting index in your DataFrame
new_df = df.reset_index()
new_df
```

```python
#15.Checking if the Stadium column is present in the DataFrame
if 'Stadium' in df.columns:
    print("'Stadium' column is present in the DataFrame.")
else:
    print("'Stadium' column is not present in the DataFrame.")
```

```python
#16.Converting datatype to float
df['AwayTeamGoals'] = df['AwayTeamGoals'].astype(float)
df.dtypes
```

```python
#17.Removing the last 10 rows from your DataFrame.
df.drop(df.tail(10).index, inplace=True)
df
```

```python
#18. Iterate over rows using iterrows()
for index, row in df.iterrows():
    # Access row data using row['column_name']
    print(f"Index: {index}, Row data: {row}")
```

```python
#19.Changing the order of your DataFrame columns.
desired_columns = [ 'Time ', 'HomeTeam', 'AwayTeam','Year', 'Date ', 'HomeTeamGoals',
       'AwayTeamGoals', 'Stage', 'SpecialWinConditions', 'Stadium', 'City',
       'Attendance', 'TotalGoals']
data = df[desired_columns]
data
```

```python
 #20.Deleting DataFrame row(s) whose value is 0 in the ‘HomeTeamGoals’ columns.
df[df['HomeTeamGoals'] != 0]
```

```python
#sum up nulls in our dataframe
df.isnull().sum()
```

```python
#Generating descriptive statistics of  our DataFrame. 
df.describe()
```

```python
#bar graphy of HomeTeamGoals and AwayTeamGoals
home = df['HomeTeamGoals'].head(20)
away = df['AwayTeamGoals'].head(20)
#figure size
fig = plt.figure(figsize=(10,7))
#horizontal bar plot
plt.bar(home[0:10],away[0:10])

plt.xlabel("Home Team Goals")
plt.ylabel("Away Team Goals")
plt.title("Home Team Goals vs Away Team Goals")
plt.show()
```
